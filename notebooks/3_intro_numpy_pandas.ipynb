@#Dot Product

```python
def dot_product(a,b):
    result = 0
    for i in range(len(a)):
        result = result + a[i]*b[i]
    return result
```

```python
J = np.random.randn(1000)
K = np.random.randn(1000)
print(J)
```

```python
print(dot_product(J,K))
```
