```python
# IQR

IQR = Q3-Q1
```

```python
IQR = df.mpg.quantile(0.75)-df.mpg.quantile(0.25)
IQR
```

```python
df.head()
```

```python
lower = df.mpg.quantile(0.25) - 1.5*IQR
upper = df.mpg.quantile(0.75) + 1.5*IQR
print(lower,upper)
```

```python
i = -1

for x in df["mpg"]:
    i += 1
    if x < lower or x > upper:
        print("Outlier: ",df.iloc[i].name)
```
