```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_features=1, noise=30, random_state=0)
```

```python
X.shape
```

```python
y.shape
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(X)
df.columns = ["x"]
df["y"] = y
df
```

```python
df["x"].hist()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
model2 = LinearRegression()
model3 = LinearRegression()
```

```python
model1.fit(X,y)
```

```python
y_pred = model1.predict(X)
```

```python
import numpy as np
y_mean = np.array([y.mean()]*100)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))
plt.scatter(X,y, label="Random Data")
plt.plot(X,y_pred, label="Linear Regression :D]", c="Red")
plt.xlabel("Features")
plt.ylabel("Target")
plt.title("Scatter sample plot")
plt.legend()

plt.savefig("../reports/figures/papuxd.svg")
```
